```python
filename = "inter1.csv"
fields = ["country","christians","muslims","unaffiliated","hindus","buddhists","folkReligions","other","jews","population"]
with open(filename, 'w') as csvfile:
  writer = csv.writer(csvfile)
  writer.writerow(fields)
  for indexp, rowp in df_pop.iterrows():
    for indexr, rowr in df_rel.iterrows():
      if rowp['Country/Territory']==rowr['country']:
        data1 = [rowr['country'], rowr['christians'],rowr['muslims'],rowr['unaffiliated'],rowr['hindus'],rowr['buddhists'],rowr['folkReligions'],rowr['other'], rowr['jews']]
        data1.append(rowp['2015 Population'])
        writer.writerow(data1)
```
